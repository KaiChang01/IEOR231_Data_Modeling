# 2. <a id='toc2_'></a>[The longest path problem with covariate control ](#toc0_)

Consider a graph with 4 nodes: 1, 2, 3, 4. Suppose we have 5 directed edges: 1->2, 1->3, 2->3, 2->4, 3->4, denoted by $e_1$ to $e_5$ (and also their lengths).



Now, suppose the joint distribution of $e\in\mathbb{R}^5$ follows a normal distribution $e\sim \operatorname{Normal}(\mu,\Sigma)$.
For the setting of parameters, we have the same as Discussion 1

````python
mu_vec = np.array([1.5, 2.0, 1.0, 3.5, 3.0])
Sigma = 2.0*np.eye(5)+4.0*np.ones((5,5))
print(Sigma)
````

### What's the expectation of the longest path from node 1 to 4? (Discussion 1)

For the longest path, we have $L=\max\left(e_1+e_4, e_1+e_3+e_5,e_2+e_5\right)$

### More efficient simulation by controlling covariate

We define the adjusted estimator by controling the longest path $C$ in expectation as:
$$
L(\lambda) = L - \lambda \hat{L},
$$
where the control variate is given by:
$$
\hat{L} = \sum_{i \in C} X_i - \sum_{i \in C} \mathbb{E}(X_i).
$$

The variance of the $L(\lambda)$ is:
$$
\text{Var}(L(\lambda)) = \text{Var}(L) + \lambda^2 \, \text{Var}(\hat{L}) - 2\lambda \, \text{Cov}(L, \hat{L}).
$$

To reduce the variance, we choose $\lambda^*$ that minimizes $\text{Var}(L(\lambda))$:
$$
\lambda^* = \arg\min_{\lambda} \text{Var}(L(\lambda)).
$$

The optimal value is found to be:
$$
\lambda^* = \frac{\text{Cov}(L, \hat{L})}{\text{Var}(\hat{L})}.
$$

Substituting $\lambda^*$ into the expression for the variance gives:
$$
\text{Var}(L(\lambda^*)) = \text{Var}(L) \cdot (1 - \rho^2),
$$
where
$$
\rho = \text{Corr}(L, \hat{L})
$$
is the correlation between $L$ and $\hat{L}$.

This shows that the variance is reduced by a factor of $(1 - \rho^2)$.

#### Simulations

````python
import numpy as np
# Original function without control variates.
def simulate_calculate_longest_path(mu_vec, Sigma, T):
    paths = np.random.multivariate_normal(mu_vec, Sigma, T)
    results = np.vstack([
        np.exp(paths[:, 0]) + np.exp(paths[:, 3]),
        np.exp(paths[:, 0]) + np.exp(paths[:, 2]) + np.exp(paths[:, 4]),
        np.exp(paths[:, 1]) + np.exp(paths[:, 4])
    ]).T
    longest_paths = np.max(results, axis=1)
    return longest_paths, results
````

````python
# Modified function with control variate based on the "longest" path in expectation.
def simulate_calculate_longest_path_with_control(mu_vec, Sigma, T, lambda_param=None):
    # Compute the expected value for each candidate path.
    expected_results = np.array([
        np.exp(mu_vec[0]+0.5*Sigma[0,0]) + np.exp(mu_vec[3]+0.5*Sigma[3,3]),           # E[X^(1)]
        np.exp(mu_vec[0]+0.5*Sigma[0,0])+np.exp(mu_vec[2]+0.5*Sigma[2,2]) + np.exp(mu_vec[4]+0.5*Sigma[4,4]), # E[X^(2)]
        np.exp(mu_vec[1]+0.5*Sigma[1,1]) + np.exp(mu_vec[4]+0.5*Sigma[4,4])          # E[X^(3)]
    ])
    
    # Determine the best candidate (largest in expectation).
    best_index = np.argmax(expected_results)
    mu_control = expected_results[best_index]
    
    # Simulate the paths.
    paths = np.random.multivariate_normal(mu_vec, Sigma, T)
    results = np.vstack([
        np.exp(paths[:, 0]) + np.exp(paths[:, 3]),
        np.exp(paths[:, 0]) + np.exp(paths[:, 2]) + np.exp(paths[:, 4]),
        np.exp(paths[:, 1]) + np.exp(paths[:, 4])
    ]).T
    
    # Original estimator (the maximum across candidates).
    longest_paths = np.max(results, axis=1)
    
    # Control variate: using the candidate that is best in expectation.
    Y = results[:, best_index]
    
    # If lambda is not provided, estimate the optimal lambda.
    if lambda_param is None:
        cov_L_Y = np.cov(longest_paths, Y, bias=True)[0, 1]
        var_Y = np.var(Y, ddof=0)
        lambda_param = cov_L_Y / var_Y if var_Y > 0 else 0
    
    # Adjust the estimator using the control variate.
    adjusted_longest_paths = longest_paths - lambda_param * (Y - expected_results[best_index]) ## since in simulation, actually we don't know the expectation of Y, so we use the sample mean of Y.
    
    return adjusted_longest_paths, lambda_param
````

````python
## Still use the example in Discussion 1
mu_vec = np.array([1.5, 2.0, 1.0, 3.5, 3.0])
Sigma = 2.0*np.eye(5)+4.0*np.ones((5,5))
print(Sigma)
T = 10000                          # Number of simulation runs

# Run original simulation.
original_longest_paths, _ = simulate_calculate_longest_path(mu_vec, Sigma, T)

# Run simulation with control variate.
adjusted_longest_paths, lambda_used = simulate_calculate_longest_path_with_control(mu_vec, Sigma, T)

# Compute expectations and variances.
mean_original = np.mean(original_longest_paths)
var_original = np.var(original_longest_paths, ddof=1)

mean_adjusted = np.mean(adjusted_longest_paths)
var_adjusted = np.var(adjusted_longest_paths, ddof=1)

print("Original estimator: mean = {:.4f}, variance = {:.4f}".format(mean_original, var_original))
print("Adjusted estimator: mean = {:.4f}, variance = {:.4f}, lambda used = {:.4f}".format(mean_adjusted, var_adjusted,lambda_used))
print("Variacne reduction fraction: {:.4f}".format((var_original - var_adjusted)/var_original))
````

````python
from scipy.stats import norm

# Plot histograms for both estimators.
plt.figure(figsize=(8, 5))
plt.hist(original_longest_paths, bins=50, density=True, alpha=0.6, label="Original Estimator", edgecolor='black')
plt.hist(adjusted_longest_paths, bins=50, density=True, alpha=0.6, label="Adjusted Estimator", edgecolor='black')



plt.title("Histogram of Original and Adjusted Estimators")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
````

What if we change the distribution to other dist?

````python
import numpy as np

# Set the mean vector for the 5 exponential components.
mu_vec = np.array([1.0, 1.0, 3.0, 1.0, 1.0])  # You can change these means if desired.

# Number of simulation runs.
T = 100000

# Compute the expected candidate values.
expected_results = np.array([
    mu_vec[0] + mu_vec[3],           # E[X^(1)]
    mu_vec[0] + mu_vec[2] + mu_vec[4], # E[X^(2)]
    mu_vec[1] + mu_vec[4]              # E[X^(3)]
])

# Determine the best candidate (largest in expectation).
best_index = np.argmax(expected_results)
mu_control = expected_results[best_index]

# Simulate the exponential components.

paths = np.column_stack([np.random.exponential(scale=mu, size=T) for mu in mu_vec])

# Form the candidate path values.
results = np.vstack([
    paths[:, 0] + paths[:, 3],           # Candidate 1
    paths[:, 0] + paths[:, 2] + paths[:, 4], # Candidate 2
    paths[:, 1] + paths[:, 4]              # Candidate 3
]).T

# Original estimator: maximum of the three candidate paths.
L = np.max(results, axis=1)

# Define the control variate: use the best candidate (largest in expectation).
Y = results[:, best_index]

# Estimate the optimal lambda if not provided.
# Note: Var(Y) can be estimated from the sample.
cov_L_Y = np.cov(L, Y, bias=True)[0, 1]
var_Y = np.var(Y, ddof=0)
lambda_opt = cov_L_Y / var_Y if var_Y > 0 else 0

# Compute the control variate adjusted estimator.
L_adj = L - lambda_opt * (Y - mu_control)

# Compute sample means and variances.
mean_original = np.mean(L)
var_original = np.var(L, ddof=1)

mean_adjusted = np.mean(L_adj)
var_adjusted = np.var(L_adj, ddof=1)

print("Exponential case with three candidate paths and control variate based on the longest path in expectation:")
print("Original estimator: mean = {:.4f}, variance = {:.4f}".format(mean_original, var_original))
print("Adjusted estimator: mean = {:.4f}, variance = {:.4f}".format(mean_adjusted, var_adjusted))
print("Optimal lambda used: {:.4f}".format(lambda_opt))
````

````python
from scipy.stats import norm

# Plot histograms for both estimators.
plt.figure(figsize=(8, 5))
plt.hist(L, bins=50, density=True, alpha=0.6, label="Original Estimator", edgecolor='black')
plt.hist(L_adj, bins=50, density=True, alpha=0.6, label="Adjusted Estimator", edgecolor='black')

# Fit normal distributions and overlay.
for data, color, label in [(L, 'b', "Original"), (L_adj, 'r', "Adjusted")]:
    mu, std = norm.fit(data)
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, mu, std)
    plt.plot(x, p, color, linewidth=2, label=f"{label} Fit (μ={mu:.2f}, σ={std:.2f})")

plt.title("Histogram of Original and Adjusted Estimators")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
````
